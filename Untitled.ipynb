# GeoJSON 사용해서 서울지역 시각화

```python
geo_path=f'data/part3_seoul_municipalities_geo_simple.json'
import json
geo_json=json.load(open(geo_path,encoding='utf-8'))
geo_json
```

```python
geo_json['features'][0]['properties']['name']
```

```python
m=folium.Map(location=[lat,long],zoom_start=12, tiles='stamen toner')
```

```python
df_cafe_vs=df_cafe.groupby(['시군구명','브랜드명'])['상호명'].count()
# df_cafe_vs['매장수비교']=df_cafe_vs['스타벅스'] - df_cafe_vs['이디야']>0
df_cafe_vs=df_cafe_vs.reset_index()
df_cafe_vs.columns=['구','브랜드명','매장수']
df_cafe_vs
```

```python
df_vs=df_cafe_vs.pivot(index='구',columns='브랜드명',values='매장수')
```

```python
df_vs['매장수비교']= df_vs['스타벅스'] - df_vs['이디야'] >0
df_vs
```

```python
# 숫자로 바꾸기
df_vs['매장수비교']=df_vs['매장수비교'].astype(float)
df_vs
```

```python
df_vs=df_vs.reset_index()
df_vs
```

```python
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df_vs,
    columns=['구','매장수비교'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='count'
).add_to(m)
```

## 이제 위경도 갖고와서 이 위에 점 찍어보자

```python
df_cafe
# 스타벅스, 이디야 df_vs에 위경도 합치기 ==> df_vs 기준으로 df_cafe의 위경도 합치기
# 피봇으로!!
```

```python
# 값들 평균내서 갖고오도록 피봇 ===> 평균해서 갖고와야 하니까 피봇 테이블 사용!!
df_gu_mean=df_cafe.pivot_table(index='시군구명',values=['경도','위도'],aggfunc='mean')
df_gu_mean=df_gu_mean.reset_index()
df_gu_mean
```

```python
# 구랑 시군구 기준으로 합치기 (df_vs.merge : df_vs를 기준으로 df_gu_mean 합칠거야 어떻게? inner로(키 겹치는 것만) 합칠거고 left는 기준이 되는 테이블의 기준 컬럼, right는 합쳐지는 얘의 컬럼)
df_vs = df_vs.merge(df_gu_mean, how='inner', left_on="구", right_on="시군구명")
df_vs
```

```python


for i in df_vs.index:
    sub_long=df_vs.loc[i,'경도']
    sub_lat=df_vs.loc[i,'위도']
    
    cafes=['스타벅스','이디야']
    for cafe in cafes:
        cafe_count=df_vs.loc[i,cafe]
        radius=np.sqrt(cafe_count)*3
        
        if cafe=='이디야':
            color='blue'
            sub_long = sub_long +0.01
        else:
            color='green'
        
        folium.CircleMarker(location=[sub_lat, sub_long], radius=radius, color=color,fill=True).add_to(m)
        # 위도, 경도 자리 조심해라!  
    
```
